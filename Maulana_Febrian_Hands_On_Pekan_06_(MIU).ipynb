### Definite Loop dengan FOR

`for` digunakan untuk mengulang aksi tertentu sejumlah yang telah ditentukan

```python
print("Kenalan yuk...")
for i in range (5):  # perhatikan bahwa di sini variabel i awalnya berisi 0 dan looping akan terus berjalan sampai i >= 5
  nm = input ("Masukkan nama:")
  print("Assalamu alaikum", nm)
print("Selesai Kenalan")
```

### Contoh Program Hitung Total Belanja

Simak video penjelasan dan buat program Hitung Total Belanja seperti yang diajarkan di video, coba eksekusi program dan pahami tiap baris source code nya

```python
# Jawaban anda di sini
print("selamat datang pelanggan")
jmlitem = 0 #variabel untuk menampung total item dibeli
totbel = 0 #variabel untuk menampung total belanja
pilih ='' #variabel untuk menampung pilihan y/t
while pilih != 't':
    namaBarang= input("masukkan nama barang: ")
    hargaBarang= int(input("masukkan harga barang: Rp. "))
    jmlhBarang = int(input("masukkan jumlah barang: Rp. "))

    jmlitem = jmlitem + jmlhBarang
    totbel = totbel + (hargaBarang*jmlhBarang)

    pilih = input("ada barang lagi? (y/t): ")

print (f"total item {jmlitem}")
print(f"total belanja = Rp.{totbel}")




```
